```python
# import 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', 120)
```

```python
# 여기서는 일반적으로 사용하는 CSV파일이 아닌 HDF5 파일이 사용되었다.
with pd.HDFStore("../input/train.h5", "r") as train:
    df = train.get("train")
```

```python
# 대략적인 데이터 형태와 변수들에 대하여 살펴보자.
df.head()
```

거의 모든 coulmn에서 Nan value가 존재한다는 것을 알 수 있다.
이러한 missing values에 대하여 어떻게 처리할지를 결정하기 위해
각 column마다 어느 정도의 Nan value가 존재하는지 파악해보자.

```python
labels = []
values = []
for col in df.columns:
    labels.append(col)
    values.append(df[col].isnull().sum())
    print(col, values[-1])
```

```python
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(12,50))
rects = ax.barh(ind, np.array(values), color = 'y')
ax.set_yticks(ind+((width)/2.))
ax.set_yticklabels(labels, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

Fundamental_5 가장 많은 missing values를 가지고 있고 대부분의 columns들이 많은 수의 missing values를 가지고 있는것을 알 수 있다.

## Correlation coefficient plot:
각 변수들과 target value와의 상관계수를 살펴보도록 하자. 


각 변수와 target value간의 상관계수를 구하기 위해서는 각 columns에 missing values가 있는 상태에서는 진행하기 어렵다.
때문에 임의적으로 missing values를 각 column의 평균값으로 대체한다.

```python
mean_values = df.mean(axis=0)
df.fillna(mean_values, inplace=True)
df.head()
```

이제 각 column의 상관계수(coefficient)를 구해보자.

```python
x_cols = [col for col in df.columns if col not in ['id','timestamp','y']]

labels = []
values = []
for col in x_cols:
    labels.append(col)
    values.append(np.corrcoef(df[col].values, df.y.values)[0,1])
    
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(12,50))
rects = ax.barh(ind, np.array(values), color='y')
ax.set_yticks(ind+((width)/2.))
ax.set_yticklabels(labels, rotation='horizontal')
ax.set_xlabel("Correlation coefficient")
ax.set_title("Correlation coefficient")
#autolabel(rects)
plt.show()
```

이를 통해 아래 4가지 변수가 가장 의미있는 상관계수를 가진다는 것을 알게되었다.
- technical_30
- technical_20
- fundamental_11
- technical_19

## Timestamp
일부 column들이 시간에 따라서 변화 할 수있다는 것을 파악하였으니 실제 시간과 관련된 'timestamp' column이 어떻게 변화하는지 살펴보자

```python
fig = plt.figure(figsize=(12, 6))
sns.countplot(x='timestamp', data=df)
plt.show()
```

'timestamp' column은 값이 늘어나면서 count되는 값도 점점 늘어나고있다는 것을 알게되었다. 또한 점체적으로는 점차적으로 변화한 형태이지만 어느 정도의 cycle을 가지고 한번에 급격히 변화하는 형태를 가지고 있다.

```python
print(len(df.id.unique()))
print("-------------------")
print(len(df.timestamp.unique()))
```

data 설명을 살펴보면 'id' column은 각각의 assets들을 나타낸다. 전체 assets 개수는 1,424개, timestamp의 개수는 1,100개 임을 알 수 있다. 이는 'timestamp'의 변화에 따라 assets은 새로 생기거나 없어지면서 변화했을 것이란걸 알 수 있다. 

이제 assets과 target value 'y'와의 관계를 살펴보도록 하자.
먼저 'y'에 대해 가장 낮은 평균값을 가진 assets를 보도록 하자.

```python
temp_df = df.groupby('id')['y'].agg('mean').reset_index().sort_values(by='y')
temp_df.head()
```

```python
id_to_use = [1431, 93, 882, 1637, 1118]
fig = plt.figure(figsize=(8, 25))
plot_count = 0
for id_val in id_to_use:
    plot_count += 1
    plt.subplot(5, 1, plot_count)
    temp_df = df.ix[df['id'] == id_val,:]
    plt.plot(temp_df.timestamp.values, temp_df.y.values)
    plt.plot(temp_df.timestamp.values, temp_df.y.cumsum())
    plt.title("Asset ID : " +str(id_val))
    
plt.show()
```

파란색 라인은 시간 흐름에 따른 'y' value 값의 변화를 나타내고 초록색은 이 값의 누적값을 나타낸다.
5개 중에 4개의 asset은 time stamp의 마지막 값인 1812까지 나타나지 않고 중간에 없어지는 것을 볼 수 있다.

이제 'y' value값과 가장 positive값을 나타내는 assets들을 살펴보도록 하자.

```python
temp_df = df.groupby('id')['y'].agg('mean').reset_index().sort_values(by='y')
temp_df.tail()
```

```python
id_to_use = [767, 226, 824, 1809, 1089]
fig = plt.figure(figsize=(8, 25))
plot_count = 0
for id_val in id_to_use:
    plot_count += 1
    plt.subplot(5, 1, plot_count)
    temp_df = df.ix[df['id']==id_val,:]
    plt.plot(temp_df.timestamp.values, temp_df.y.values)
    plt.plot(temp_df.timestamp.values, temp_df.y.cumsum())
    plt.title("Asset ID : " + str(id_val))
plt.show()
```

여기에서도 마찬가지로 time stamp의 마지막 값이 1812 이전에 drop 되는 assets(Assets 824, 1089)를 볼 수 있다.
